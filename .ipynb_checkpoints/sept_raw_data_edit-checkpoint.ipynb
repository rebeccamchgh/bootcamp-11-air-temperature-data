```python
# Create a function for files so that each is by day and night 5 am to 10 pm for HOBO sensors
def day_night_temps(df, sensor_number):
    df.index = pd.to_datetime(df.index)
    df['Time of Day'] = 'Day'
    df.loc[(df.index.hour >= 22) | (df.index.hour <= 6), 'Time of Day'] = 'Night'
    df['sensor_number'] = sensor_number
    return df
```

```python
 # Function to edit kestrel sensors
def import_edit_irrigation(file_path):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)

    # Select the desired columns and rename them
    df = df.iloc[:, :3]  # Keep the first 3 columns
    df.columns = ['date_time', 'temp_f', 'humidity']  # Rename the columns

    # Drop any rows with missing values in 'date_time' and 'temp_f' columns
    df.dropna(subset=['date_time', 'temp_f'], inplace=True)

    # Convert 'date_time' column to datetime format
    df['date_time'] = pd.to_datetime(df['date_time'])

    return df
```

```python
# Function for 12 hour 
import pandas as pd

def process_csv(file_path):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)

    # Check if the DataFrame has at least 3 columns
    if len(df.columns) >= 3:
        # Keep only the first three columns
        df = df.iloc[:, :3]

        # Rename the columns to 'date_time', 'temp_f', and 'humidity'
        df.columns = ['date_time', 'temp_f', 'humidity']

        # Drop any rows with missing values in 'date_time' and 'temp_f' columns
        df.dropna(subset=['date_time', 'temp_f'], inplace=True)

    return df
```

```python
# Convert to 24 hour time 
sensor_044a2 = process_csv('temperature-data-sept-2023/#044A_Aug_22_2023_18_05_00.csv')
sensor_044a2 = sensor_044a2.drop(0)
sensor_044b2 = process_csv('temperature-data-sept-2023/#044B_Aug_21_2023_01_45_00.csv')
sensor_044c2 = process_csv('temperature-data-sept-2023/#044C_Aug_10_2023_01_30_00.csv')
sensor_045a2 = process_csv('temperature-data-sept-2023/#045A_Aug_22_2023_15_25_00.csv')
sensor_045b2 = process_csv('temperature-data-sept-2023/#045B_Aug_8_2023_02_35_00.csv')
sensor_045c2 = process_csv('temperature-data-sept-2023/#045C_Aug_22_2023_15_40_00.csv')
sensor_046a2 = process_csv('temperature-data-sept-2023/#046A_replacement_Aug_22_2023_11_10_00.csv')
# edit 46a2
#sensor_046b1 = import_edit_irrigation( 'temperature-data-sept-2023/#046B_replacement_Jul_29_2023_23_15_00.csv')
# check replacment date 46b2
sensor_047a2 = process_csv('temperature-data-sept-2023/#047A_Aug_31_2023_9_10_00_PM.csv')
sensor_047b2 = process_csv('temperature-data-sept-2023/#047B_Aug_22_2023_14_40_00.csv')
sensor_048a2 = process_csv('temperature-data-sept-2023/#048A_Aug_22_2023_16_45_00.csv')
sensor_048b2 = process_csv('temperature-data-sept-2023/#048B_Aug_22_2023_17_05_00.csv')
sensor_049a2 = process_csv('temperature-data-sept-2023/#049A_Aug_22_2023_11_50_00.csv')
# edit dates here for 49a2
#sensor_049b = 
```

```python

# Create a list of the DataFrames and their corresponding sensor names
sensor_data = [
    (sensor_044a2, 'sensor_044a'),
    (sensor_044b2, 'sensor_044b'),
    (sensor_044c2, 'sensor_044c'),
    (sensor_045a2, 'sensor_045a'),
    (sensor_045b2, 'sensor_045b'),
    (sensor_045c2, 'sensor_045c'),
    (sensor_046a2, 'sensor_046a'),
    (sensor_047a2, 'sensor_047a'),
    (sensor_047b2, 'sensor_047b'),
    (sensor_048a2, 'sensor_048a'),
    (sensor_048b2, 'sensor_048b'),
    (sensor_049a2, 'sensor_049a'),
    # Add more DataFrames and sensor names as needed
]

# Concatenate the DataFrames into one while adding a 'Sensor' column
irrigation_august = pd.concat([df.assign(Sensor=sensor_name) for df, sensor_name in sensor_data], ignore_index=True)

irrigation_august['date_time'] = pd.to_datetime(irrigation_august['date_time'])

def classify_time(row):
    if row.time() >= pd.Timestamp('2023-07-27 07:00:00').time() and row.time() <= pd.Timestamp('2023-08-28 22:00:00').time():
        return "day"
    else:
        return "night"

# Add a new column 'time_of_day' to the DataFrame
irrigation_august['time_of_day'] = irrigation_august['date_time'].apply(classify_time)
irrigation_august
```

```python
#  Convert 'temp_f' to numeric (float) type
irrigation_august['temp_f'] = pd.to_numeric(irrigation_august['temp_f'], errors='coerce')

# Convert 'date_time' to datetime
irrigation_august['date_time'] = pd.to_datetime(irrigation_august['date_time'])

# Extract 'day' or 'night' based on the time
#irrigation_august['time_of_day'] = irrigation_august['date_time'].dt.hour.apply(lambda x: 'day' if 7 <= x < 22 else 'night')

# Group by 'Sensor', 'time_of_day', and calculate daily mean temperature
temp_mean = irrigation_august.groupby(['Sensor', 'time_of_day', irrigation_august['date_time'].dt.date])['temp_f'].mean().reset_index()

# Rename the column to indicate it's the mean temperature
temp_mean.rename(columns={'temp_f': 'mean_temp'}, inplace=True)

# Print the daily mean temperature for each sensor and each time of day
temp_mean

```

```python
# Convert 'humidity' column to numeric (if it's not already)
irrigation_august['humidity'] = pd.to_numeric(irrigation_august['humidity'], errors='coerce')

# Group by 'Sensor', 'time_of_day', and calculate daily mean humidity
humidity_mean = irrigation_august.groupby(['Sensor', 'time_of_day', irrigation_august['date_time'].dt.date])['humidity'].mean().reset_index()

# Rename the column to indicate it's the mean humidity
humidity_mean.rename(columns={'humidity': 'mean_humidity'}, inplace=True)

# Print the daily mean humidity for each sensor and each time of day
humidity_mean

august = temp_mean.merge(humidity_mean, on=['Sensor', 'time_of_day', 'date_time'])

# Print the merged DataFrame
august['date_time'] = pd.to_datetime(august['date_time'])

august
```

```python
# Define the date range you want to plot
start_date = '2023-07-30'
end_date = '2023-08-20'

# Define the sensors you want to plot
selected_sensors = ['sensor_044a', 'sensor_044b']

# Filter data for the specified date range, selected sensors, and daytime records
filtered_df = august[(august['date_time'] >= start_date) & (august['date_time'] <= end_date) & 
                     (august['Sensor'].isin(selected_sensors)) & (august['time_of_day'] == 'day')]

# Create a plot for daytime temperature and humidity
plt.figure(figsize=(12, 6))

for sensor in filtered_df['Sensor'].unique():
    sensor_data = filtered_df[filtered_df['Sensor'] == sensor]
    sorted_data = sensor_data.sort_values(by='date_time')  # Sort by 'date_time'
    plt.plot(sorted_data['date_time'], sorted_data['mean_temp'], label=f'{sensor} Daytime Temperature', marker='o')
    plt.plot(sorted_data['date_time'], sorted_data['mean_humidity'], label=f'{sensor} Daytime Humidity', marker='o', linestyle='--')

plt.xlabel('Date')
plt.ylabel('Relative Humidity (%) and Mean temperature (F)')
plt.title('Daytime Temperature and Humidity')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Add day/night for kestrel sensors. 
def add_time_of_day_and_sensor_number(df, sensor_number):
    df['date_time'] = pd.to_datetime(df['date_time'])
    df['Time of Day'] = 'Day'
    df.loc[(df['date_time'].dt.hour >= 22) | (df['date_time'].dt.hour <= 6), 'Time of Day'] = 'Night'
    df['sensor_number'] = sensor_number
    return df
```

```python
# Edit start dates for kestrel sensors
def filter_data_for_sensor(df, sensor_number, start_date):
    df['date_time'] = pd.to_datetime(df['date_time'])
    filtered_df = df[(df['date_time'] >= start_date) & (df['sensor_number'] == sensor_number)]
    return filtered_df
```
